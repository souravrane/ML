```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
x = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)
```

```python
x.columns = ['sl','sw','pl','pw']
y.columns = ['target']
colormap = np.array(['red','blue','orange'])
```

```python
#plot for sepals
import matplotlib.pyplot as plt
plt.scatter(x.sl,x.sw,c=colormap[y.target])
plt.show()
```

```python
#plot for petals
plt.scatter(x.pl,x.pw,c=colormap[y.target])
plt.show()
```

```python
#Kmeans
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
km.fit(x)
```

```python
#after kmeans
plt.scatter(x.pl,x.pw,c=colormap[km.labels_],s=40)
plt.show()
from sklearn import metrics
print(metrics.accuracy_score(y,km.labels_))
```

```python
#EM
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=3)
y_pred = gm.fit(x).predict(x)
plt.scatter(x.pl,x.pw,c=colormap[y_pred])
plt.show()
```
